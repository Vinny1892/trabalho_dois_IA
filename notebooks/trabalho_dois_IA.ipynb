```python
wifis = pd.read_csv('../data/dataset.csv')
x = wifis.drop('classe', axis=1)
x = x.drop(['FARMA POPULAR','#NET-CLARO-WIFI','Unnamed: 30','moto g(6) plus 3609', 'Redmi Note 9S','70mai_Pro Plus+_8eb4'] ,axis=1)
y = wifis['classe']
len(x.columns)
```

```python
y = wifis['classe']
print('claase' not in x.columns)
print(x.columns)
print(x.shape[1])
```

```python
std_sacler = StandardScaler()
x_normalize = std_sacler.fit_transform(x)
print("Dados normalizados com sucesso")
```

```python
 ## Fase 3 (peso 2): Treino e ajuste de parâmetros.  Utilizando o conjunto de validação, encontre o melhor  conjunto de parâmetros para o seu problema. Vc pode utilizar a quantidade de vizinhos, diferentes valores de p, pesos, etc.
```

```python
sss = StratifiedShuffleSplit(n_splits=10, test_size=0.30, train_size=0.7, random_state=0)
lacc = []
melhores_vizinhos = {}
for k in range(2,19):
    model = KNeighborsClassifier(n_neighbors=k)
    for train_index,test_index in sss.split(x_normalize,y):
        xtrain = x_normalize[train_index]
        ytrain = y[train_index]
        xtest  = x_normalize[test_index]
        ytest  = y[test_index]
        x_test,x_validation ,y_test , y_validation = train_test_split(xtest, ytest, test_size=0.5, random_state=0)
        model.fit(x_validation,y_validation)
        ypred=model.predict(x_test)
        acc = accuracy_score(y_test,ypred)
        print("Precision Score : ",precision_score(y_test,ypred, average='micro'))
        print("Recall Score : ",recall_score(y_test,ypred,average='micro'))
        print("f1 Score : ",f1_score(y_test,ypred,average='micro'))
        print(f'{confusion_matrix(y_test,ypred)}')
        print(f'accuracy_score {acc} para numero de vizinhos {k} ')
        lacc.append(acc)
        media = np.mean(lacc)
        desvio_padrao = np.std(lacc)
        f1_score_para_k = f1_score(y_test,ypred,average='micro')
    melhores_vizinhos[f1_score_para_k] = (k, desvio_padrao, media)
    print("qtd_vizinhos: %2d , Média: %4.3f Std: %4.3f " % (k,media, desvio_padrao) )
    f1_melhor_vizinho = max(melhores_vizinhos)
    melhor_vizinho = melhores_vizinhos[f1_melhor_vizinho]
    print('melhor quantidade de vizinhos:',melhor_vizinho[0],',media:', melhor_vizinho[2],',desvio padrao:',melhor_vizinho[1])
    print(f"Média: %4.3f Std: %4.3f para {k} vizinhos "%(np.mean(lacc),np.std(lacc)))
    print('_____________________________________________________________________________________________________________________________________')
```

## PARA MELHOR NUMERO DE VIZINHOS ACAHDOS ANTERIORMENTE NA CELULA ACIMA


```python
print("PARA MELHOR NUMERO DE VIZINHOS ACAHDOS ANTERIORMENTE NA CELULA ACIMA")
sss = StratifiedShuffleSplit(n_splits=10, test_size=0.30, random_state=0)
lacc = []
model = KNeighborsClassifier(n_neighbors=melhor_vizinho[0])
for train_index,test_index in sss.split(x_normalize,y):
    xtrain = x_normalize[train_index]
    ytrain = y[train_index]
    xtest  = x_normalize[test_index]
    ytest  = y[test_index]
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    acc = accuracy_score(ytest,ypred)
    print("Precision Score : ",precision_score(ytest,ypred, average='micro'))
    print("Recall Score : ",recall_score(ytest,ypred,average='micro'))
    print("f1 Score : ",f1_score(ytest,ypred,average='micro'))
    print(confusion_matrix(ytest,ypred))
    print(f'accuracy_score {acc} para numero de vizinhos {melhor_vizinho[0]} ')
    lacc.append(acc)
print(f"Média: %4.3f Std: %4.3f para {melhor_vizinho[0]} vizinhos "%(np.mean(lacc),np.std(lacc)))
```
